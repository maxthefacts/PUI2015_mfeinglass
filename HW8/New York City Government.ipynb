```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
from colour import Color
import requests as r
%pylab inline 
```

Grab data from github

```python
url = 'https://raw.githubusercontent.com/maxthefacts/PUI2015_mfeinglass/master/HW8/nyc.csv'
data = pd.read_csv(url)
```

Create Labels of Department Names

```python
labels = data['AGENCY EXPENDITURES']
arraylabels = labels.as_matrix()
data = data[data.columns[1:]]
data = data.dropna()
arraylabels = arraylabels.astype('string')
arraylabels = np.fliplr([arraylabels])[0]
```

Format for Stacked Graph

```python
y = data.as_matrix()
x = [2013, 2012, 2011, 2010, 2009, 2008, 2007, 2006, 2005, 2004, 2003, 2002, 2001]
```

Color ramp created from color brew using 'Color' package

```python
red = Color("red")
colors = list(red.range_to(Color("green"),22))
colors = colors[1:-1]
colors1 = ['#f32e00', '#ed4400', '#e75800', '#e16b00', '#db7d00', '#d58e00', '#cf9d00', '#c9ac00', '#c3b900', '#b4bc00', '#9cb600', '#86b000', '#72aa00', '#5ea400', '#4b9e00', '#3a9800', '#2a9200', '#1b8c00', '#0d8600']
```

Create Plot with Legend

```python
fig = plt.figure(figsize = (14,10))
ax = fig.add_subplot(111)

ax.stackplot(x, y, label = arraylabels, baseline = 'weighted_wiggle', colors = colors1)
ax.set_title('New York City Department Budgets 2001 to 2013', size = 20)
ax.set_xlabel('Year', size = 20)
ax.set_ylabel('Percent of NYC Budget', size = 20)
ax.set_yticklabels([])
ax.margins(0, 0) # Set margins to avoid "whitespace"
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

plt.legend([mpatches.Patch(color=colors1[0]),  
            mpatches.Patch(color=colors1[1]),
            mpatches.Patch(color=colors1[2]),
            mpatches.Patch(color=colors1[3]),
            mpatches.Patch(color=colors1[4]),
            mpatches.Patch(color=colors1[5]),
            mpatches.Patch(color=colors1[6]),
            mpatches.Patch(color=colors1[7]),
            mpatches.Patch(color=colors1[8]),
            mpatches.Patch(color=colors1[9]),
            mpatches.Patch(color=colors1[10]),
            mpatches.Patch(color=colors1[11]),
            mpatches.Patch(color=colors1[12]),
            mpatches.Patch(color=colors1[13]),
            mpatches.Patch(color=colors1[14]),
            mpatches.Patch(color=colors1[15]),
            mpatches.Patch(color=colors1[16]),
            mpatches.Patch(color=colors1[17]),
            mpatches.Patch(color=colors1[18])],
            arraylabels, bbox_to_anchor=(1.05, 1), loc=2,)
plt.show()
```
